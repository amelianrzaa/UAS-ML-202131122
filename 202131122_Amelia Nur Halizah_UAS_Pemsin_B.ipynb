# Kurva bobot perbandingan clustering untuk n-cluster dari 1 - 7

```python
# hitung within cluster square sum of error (WCSS) untuk setipa cluster yang berbeda

from sklearn.cluster import KMeans

wcss = []
for i in range(1, 8):
    kModel = KMeans(n_clusters=i, n_init=10)
    kModel.fit(df[['horsepower', 'mpg']])
    wcss.append(kModel.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 8), wcss)
plt.plot(range(1, 8), wcss, color='red', linewidth=2, marker='8')
plt.xticks(np.arange(1, 8))
plt.ylabel("WCSS")
plt.show()
```

gambar diatas adalah kurva daya mobil dengan konsumsi bahan bakar

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 8):
    kModel = KMeans(n_clusters=i, n_init=10)
    kModel.fit(df[['weight', 'mpg']])
    wcss.append(kModel.inertia_)
plt.figure(figsize=(10, 6))
plt.plot(range(1, 8), wcss)
plt.plot(range(1, 8), wcss, color='red', linewidth=2, marker='8')
plt.xticks(np.arange(1, 8))
plt.ylabel("WCSS")
plt.show()
```

gambar diatas adalah kurva bobot mobil dengan konsumsi bahan bakar

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 8):
    kModel = KMeans(n_clusters=i, n_init=10)
    kModel.fit(df[['displacement', 'mpg']])
    wcss.append(kModel.inertia_)
plt.figure(figsize=(10, 6))
plt.plot(range(1, 8), wcss)
plt.plot(range(1, 8), wcss, color='red', linewidth=2, marker='8')
plt.xticks(np.arange(1, 8))
plt.ylabel("WCSS")
plt.show()
```

# menggunakan model k-means dengan n cluster 2 yang di mana, di dapat dari grafik diatas bahwa titik siku berada pada iterasi n-cluster ke 2

```python
model = KMeans(n_clusters=2, n_init=10)
model.fit(df[['horsepower', 'mpg']])
prediction = model.predict(df[['horsepower', 'mpg']])
prediction
```

```python
df['Label'] = prediction
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='horsepower', y='mpg', hue='Label',
                palette=['green', 'blue',], data=df, legend='full', s=50)
plt.title('Sebaran Data')
plt.show()
```
